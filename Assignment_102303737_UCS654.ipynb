```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('data.csv', encoding='latin-1')
```

```python
x = df['no2'].dropna().values
```

```python
r = 102303723
# Computing ar and br
print("r % 7 =", r % 7)
print("r % 5 =", r % 5)

ar = 0.05 * (r % 7)
br = 0.3 * (r % 5 + 1)

ar, br
```

```python
# Apply transformation Step 1
z = x + ar * np.sin(br * x)
```

```python
# Step 2
# Calculating mean
mu = np.mean(z)
# Calculating Lambda variance
variance = np.mean((z - mu) ** 2)
lambda_ = 1 / (2 * variance)
# c is normalization constant
c = np.sqrt(lambda_ / np.pi)
print("mu =", mu)
print("lambda =", lambda_)
print("c =", c)
```

```python
z_shuffled = np.random.permutation(z)
mu2 = np.mean(z_shuffled)
var2 = np.mean((z_shuffled - mu2)**2)
lambda2 = 1 / (2 * var2)
c2 = np.sqrt(lambda2 / np.pi)

mu2, lambda2, c2
```
